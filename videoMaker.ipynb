```python
import cv2
import numpy as np
import glob
import os

thisPath=os.getcwd()

img_array = []

pathlist =glob.glob(thisPath+'\\them2\\*.png')
temolist=[]
for path in pathlist:
    temolist.append(int(path.split("fram ")[1].split(".")[0]))


pathlist= [x for _, x in sorted(zip(temolist, pathlist))]

print(pathlist)

```

```python
for filename in pathlist:
    print(filename)
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)
```

```python

out = cv2.VideoWriter('Theme2.avi',cv2.VideoWriter_fourcc(*'DIVX'), 24, size)
 
for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
```
